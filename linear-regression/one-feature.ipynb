```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
```

```python
global input_size, output_price

input_size = np.array([
    2, 5, 7, 9, 10 , 1, 11, 25, 14, 20, 21, 4, 8, 14
])
output_price = np.array([
    12, 25, 17, 6, 20 , 11, 21, 22, 30, 29, 35, 38, 30, 32
])

print(input_size.shape)
print(output_price.shape)
```

```python
theta_0 = random.randn()
theta_1 = random.randn()

def print_hypothesis(theta_0, theta_1):
  print('theta_0', theta_0)
  print('theta_1', theta_1)
  print(f'hypothesis = {theta_0} + {theta_1} * x')

print_hypothesis(theta_0, theta_1)
```

```python
def plot(theta_0, theta_1):
  plt.plot(input_size, output_price, 'o')

  x = np.linspace(0, 25)
  plt.plot(x, theta_0 + (theta_1 * x), linestyle = 'solid')

plot(theta_0, theta_1)
plt.show()
```

```python
# Loss function / Least Squares
# J(θ) = 1/2 * Σ((h(x) - y)^2)
# where h(x) = θ_0 + θ_1 * x
def loss_function(input, output, c, m):
  loss = np.sum((((input * m) + c) - output) ** 2) / 2
  return loss

loss_function(input_size, output_price, theta_1, theta_0)
```

```python
# Stochastic Gradient Descent
def gradient_descent(input, output, theta_0, theta_1):
  alpha = 0.0001

  for i in range(100):
    print(f'Iteration: {i},  Loss: {loss_function(input_size, output_price, theta_0, theta_1)}')
    plot(theta_0, theta_1)

    sum_theta_1 = 0
    sum_theta_0 = 0
    for j in range(len(input)):
      sum_theta_0 = sum_theta_0 + ((theta_1 * input[j] + theta_0) - output[j]) * 1
      sum_theta_1 = sum_theta_1 + ((theta_1 * input[j] + theta_0) - output[j]) * input[j]
    
    theta_0 = theta_0 - alpha * sum_theta_0
    theta_1 = theta_1 - alpha * sum_theta_1
  
  return (theta_1, theta_0)

new_theta_1, new_theta_0 = gradient_descent(input_size, output_price, theta_0, theta_1)

print_hypothesis(new_theta_0, new_theta_1)
plt.show()
```

```python
def predict_price(size):
  return (new_theta_0 + size * new_theta_1)

print(predict_price(10000))
```
